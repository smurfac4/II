4.3.1 –ó–∞–¥–∞–π—Ç–µ –º–∞—Å—Å–∏–≤ —Å–ª—É—á–∞–π–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∏–∑ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞(0; 1).
–†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω–∏–µ –∏ –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –º–∞—Å—Å–∏–≤–∞, —Å—Ä–∞–≤–Ω–∏—Ç–µ
—Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, –∫–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –æ –∑–Ω–∞—á–µ–Ω–∏—è—Ö?
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ–Ω–∏—è –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞.


```python
from random import randint
import statistics as s
from matplotlib import pyplot as plt

arr = []
arr2 = []
for i in range(100):
    f = [randint(0, 1) for i in range(100)]
    arr.append(s.mean(f)) #—Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
    arr2.append(s.median(f)) #–º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ 

plt.grid()
plt.scatter(arr,arr2) #–∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Å—Ä–µ–¥–Ω–µ–≥–æ –æ—Ç –º–µ–¥–∏–∞–Ω–Ω–Ω–æ–≥–æ 
plt.show()



```

–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì

```python
import math as np 
from random import randint
import statistics as s
from matplotlib import pyplot as plt
import numpy as np

arr = [(np.sqrt(1+np.exp(np.sqrt(x))+np.cos(x**2)))/(np.fabs(1-np.sin(x)**3)) + np.log1p(np.fabs(x*2)) for x in range(1,10)]
print (arr)

plt.grid()
plt.plot(arr[:5],marker = "p")
plt.plot(arr)
plt.show()



```

```python
import math as m 
from random import randint
import statistics as s
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(0,10,1)
y = np.abs(np.cos(x*np.exp((np.cos(x)+np.log1p(x+1)))))

plt.grid()
plt.plot(x,y,c = "b")
plt.fill_between(x,y)

area = np.trapz(y)
print(area)
```
