```python
import numpy as np 
def sigmoid(x):
    return 1 / (1 + np.exp(-x)) # Ð¤ÑƒÐ½ÐºÑ†Ð¸Ñ Ð°ÐºÑ‚Ð¸Ð²Ð°Ñ†Ð¸Ð¸: f(x) = 1 / (1 + e*(-x))

class Neuron: # Ð½ÐµÐ¹Ñ€Ð¾Ð½
    def __init__(self, weights, bias): 
        self.weights = weights
        self.bias = bias
    def feedforward(self, inputs):
        total = np.dot(self.weights, inputs) + self.bias
        return sigmoid(total)
```

```python
class OurNeuralNetwork1: #Ð´Ð°Ð½Ð½Ñ‹Ðµ Ð½ÐµÐ¹Ñ€Ð¾ÑÐµÑ‚Ð¸
    def __init__(self):
        self.h1 = Neuron(weights, bias)
        self.h2 = Neuron(weights, bias) 
        self.h3 = Neuron(weights, bias) 
        self.o1 = Neuron(weights, bias)
    def feedforward(self, x): 
        out_h1 = self.h1.feedforward(x) 
        out_h2 = self.h2.feedforward(x)
        out_h3 = self.h3.feedforward(x)
        # Ð’Ñ…Ð¾Ð´Ñ‹ Ð´Ð»Ñ o1 - ÑÑ‚Ð¾ Ð²Ñ‹Ñ…Ð¾Ð´Ñ‹ h1 u h2 Ð¸ h3
        out_o1 = self.o1.feedforward(np.array([out_h1, out_h2,out_h3])) 
        return out_o1


#Ð´Ð°Ð½Ð½Ñ‹Ðµ Ð½ÐµÐ¹Ñ€Ð¾Ð½Ð°
weights = np.array([0.5,0.5,0.5]) # ð‘¤ = [0.5, 0.5, 0.5]
bias = 0  # ð‘ = 0

network = OurNeuralNetwork1()
x = np.array([1,1,1]) # Ð¿ÑƒÑÑ‚ÑŒ x = 1, y = 1 , z = 1
print(network.feedforward(x))
```

```python
import numpy as np 

# ====Ð—Ð°Ð´Ð°Ð½Ð¸Ðµ_2====
def sigmoid1(x):
    return 1 / (1 + np.exp(-x))# Ð¤ÑƒÐ½ÐºÑ†Ð¸Ñ Ð°ÐºÑ‚Ð¸Ð²Ð°Ñ†Ð¸Ð¸: sigmoid

def sigmoid2(x):
    return np.tan(x)# Ð¤ÑƒÐ½ÐºÑ†Ð¸Ñ Ð°ÐºÑ‚Ð¸Ð²Ð°Ñ†Ð¸Ð¸: tanh

def sigmoid3(x):
    return max(0,x) # Ð¤ÑƒÐ½ÐºÐ¹Ð¸Ñ Ð°ÐºÑ‚Ð¸Ð²Ð°Ñ†Ð¸Ð¸: ReLU
# ====ÐšÐ¾Ð½ÐµÑ†_2_Ð—Ð°Ð´Ð°Ð½Ð¸Ñ====

class Neuron: # ÑÐ°Ð¼ Ð½ÐµÐ¹Ñ€Ð¾Ð½
    def __init__(self, weights, bias): 
        self.weights = weights
        self.bias = bias
    def feedforward(self, inputs):
        total = np.dot(self.weights, inputs) + self.bias
        return sigmoid3(total)
```

```python
class OurNeuralNetwork2: #Ð´Ð°Ð½Ð½Ñ‹Ðµ Ð½ÐµÐ¹Ñ€Ð¾ÑÐµÑ‚Ð¸
    def __init__(self):
        self.h1 = Neuron(weights, bias)
        self.h2 = Neuron(weights, bias) 
        self.o1 = Neuron(weights, bias)
        self.o2 = Neuron(weights, bias)
    def feedforward(self, x): 
        out_h1 = self.h1.feedforward(x) 
        out_h2 = self.h2.feedforward(x)
        # Ð’Ñ…Ð¾Ð´Ñ‹ Ð´Ð»Ñ o1 Ð¸ o2 - ÑÑ‚Ð¾ Ð²Ñ‹Ñ…Ð¾Ð´Ñ‹ h1 u h2
        out_o1 = self.o1.feedforward(np.array([out_h1, out_h2])) 
        out_o2 = self.o2.feedforward(np.array([out_h1, out_h2]))
        return out_o1,out_o2

#Ð´Ð°Ð½Ð½Ñ‹Ðµ Ð½ÐµÐ¹Ñ€Ð¾Ð½Ð°
weights = np.array([1,0]) # ð‘¤ = [1,0] 
bias = 1  # ð‘ = 1

network = OurNeuralNetwork2()
x = np.array([1,1]) # Ð¿ÑƒÑÑ‚ÑŒ x = 1, y = 1
print(network.feedforward(x))
```
